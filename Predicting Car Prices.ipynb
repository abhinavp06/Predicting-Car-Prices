```python
# Now, we're going to drop the original Price column since its no longer needed.
data_cleaned = data_cleaned.drop(['Price'],axis = 1)
```

### Mutlicollinearity

```python
data_cleaned.columns.values
```

```python
# Checking for multicollinearity using statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
```

```python
vif
```

```python
# We're gonna set the limit of max VIF = 10
# THerefore, we're going to get rid of the Year column
data_no_multicollinearity = data_cleaned.drop(['Year'] , axis=1)
```

### Creating Dummy Variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity,drop_first = True)
```

```python
data_with_dummies
```

##### Rearranging the columns

```python
data_with_dummies.columns.values
```

```python
cols = ['Log Price','Mileage','EngineV', 'Brand_BMW', 'Brand_Mercedes-Benz','Brand_Mitsubishi','Brand_Renault',
        'Brand_Toyota','Brand_Volkswagen','Body_hatch','Body_other', 'Body_sedan','Body_vagon', 'Body_van', 
        'Engine Type_Gas','Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed
```

### Linear Regression Model

```python
targets = data_preprocessed['Log Price']
inputs = data_preprocessed.drop(['Log Price'], axis = 1)
```

#### Scaling Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

##### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled,targets,test_size=0.2,random_state=365)
```

#### Creating the regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets(y_train)',size=18)
plt.ylabel('Predictions(y_hat)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# Residual Plot
sns.distplot(y_train - y_hat)
plt.title('Residual PLot',size=18)
```

```python
# Calculating R-squared
reg.score(x_train,y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
# +ve weight means, log_price (and Price) increases as the feature increases. A -ve weight means the opposite.
```

```python
data_cleaned['Brand'].unique()
```

### Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

##### Our model seems to be pretty accurate for high prices. Not so much for the lower price bracket.

```python
plt.scatter(y_test,y_hat_test,alpha=0.2)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_performance = pd.DataFrame(np.exp(y_hat_test),columns = ['Prediction']) #Taking exponent power as we get the log values without it.
df_performance
```

```python
df_performance['Target'] = np.exp(y_test)
df_performance
```

###### Unfortunately, we have a lot of missing values spread randomly

```python
y_test = y_test.reset_index(drop=True)
y_test
```

```python
df_performance['Target'] = np.exp(y_test)
df_performance
```

```python
df_performance['Residual'] = df_performance['Target'] - df_performance['Prediction']
```
